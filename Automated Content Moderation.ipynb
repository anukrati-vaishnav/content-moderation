```python
#Compute the percentage of toxic comments in the dataset.

# Total comments
total_comments = len(df_train)
print(f"Total comment: {total_comments}") #total number of rows

# Toxic comments (at least one toxicity label is 1)
df_train['is_toxic'] = df_train[toxicity_columns].sum(axis=1) > 0
toxic_comments = df_train['is_toxic'].sum()
print(f"Toxic comments: {toxic_comments}") 

# Toxicity prevalence
toxicity_prevalence = (toxic_comments / total_comments) * 100
print(f"Toxicity Prevalence: {toxicity_prevalence:.2f}%")

```

```python
#Estimate Manual Moderation and Automated Moderation Costs
#Assume a cost per comment for manual moderation and calculate the total cost for handling toxic comments manually.

# Manual moderation cost
manual_moderation_cost = 30  # $ per hour
comments_per_hour = 100
hours_needed = len(df_train) / comments_per_hour
total_manual_cost = manual_moderation_cost * hours_needed

# Automated moderation cost
ai_training_cost = 5000  # Training cost
inference_cost_per_comment = 0.01
total_ai_cost = ai_training_cost + len(df_train) * inference_cost_per_comment

# Cost savings
savings = total_manual_cost - total_ai_cost
print(f"Manual Moderation Cost: ${total_manual_cost:.2f}")
print(f"AI Moderation Cost: ${total_ai_cost:.2f}")
print(f"Cost Savings: ${savings:.2f}")
```

```python
# Visualization
categories = ['Manual Moderation', 'AI Moderation', 'Savings']
values = [total_manual_cost, total_ai_cost, savings]

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=['blue', 'green', 'orange'])
plt.title("Cost-Benefit Analysis of Automated Moderation")
plt.ylabel("Cost ($)")
plt.xlabel("Cost Categories")
for i, v in enumerate(values):
    plt.text(i, v + 5000, f"${v:,.2f}", ha='center')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python


import string
import re
from textstat import flesch_reading_ease

# data = {'comment_text': ["This is an example comment! And it is not toxic.", "This is a BAD comment!!! VERY BAD!!"]}
# df = pd.DataFrame(data)

# Feature: Text length
df_train['text_len'] = df_train['comment_text'].apply(len)

# Feature: Word count
df_train['word_count'] = df_train['comment_text'].apply(lambda x: len(x.split()))

# Feature: Unique word count
df_train['unique_word_count'] = df_train['comment_text'].apply(lambda x: len(set(x.split())))

# Feature: Punctuation count
df_train['punctuation_count'] = df_train['comment_text'].apply(lambda x: len([c for c in x if c in string.punctuation]))

# Feature: Upper case word count
df_train['upper_case_word_count'] = df_train['comment_text'].apply(lambda x: len([w for w in x.split() if w.isupper()]))

# Feature: Average word length
df_train['avg_word_length'] = df_train['comment_text'].apply(lambda x: np.mean([len(w) for w in x.split()]))

# Feature: Sentiment polarity score
df_train['sentiment_polarity'] = df_train['comment_text'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Feature: Sentiment subjectivity score
df_train['sentiment_subjectivity'] = df_train['comment_text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Feature: Readability score
df_train['readability_score'] = df_train['comment_text'].apply(lambda x: flesch_reading_ease(x))

# Feature: Profanity count (define a more robust set of profanity words based on your use case)
profanity_words = {'badword1', 'badword2'}  # Example set of profanity words, expand this list as needed
df_train['profanity_count'] = df_train['comment_text'].apply(lambda x: len([w for w in x.lower().split() if w in profanity_words]))

# Check the DataFrame to see the new features
df_train.head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load dataset

df_train = pd.read_csv("/Users/anukrativaishnav/CapstoneProject/toxic-comment-train.csv.zip")
print(df_train.columns)

# The dataset has 'comment_text' and 'toxic' columns
# 1: Toxic, 0: Non-Toxic
labeled_data = df_train[df_train['toxic'].notnull()]  # Rows with labels
unlabeled_data = df_train[df_train['toxic'].isnull()]  # Rows without labels (if any)

```

```python
#Check label distribution

print("Label Distribution in Labeled Data:")
print(labeled_data['toxic'].value_counts())
```

```python
#Logistic Regression
```

```python
#Train the initial model with the labelled data

from sklearn.feature_extraction.text import TfidfVectorizer
# Prepare labeled data
X_labeled = labeled_data['comment_text']
y_labeled = labeled_data['toxic']

# Split labeled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_labeled, y_labeled, test_size=0.2, random_state=42)


# Convert text data to TF-IDF features
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)


# Train a Logistic Regression model
model = LogisticRegression()
model.fit(X_train_vec, y_train)

#Evaluate the Model- Test the model on the test set and evaluate its performance using a confusion matrix and classification report.

# Make predictions on the test set
y_pred = model.predict(X_test_vec)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=['Non-Toxic', 'Toxic'])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix: Toxic Comment Moderation")
plt.show()

# Classification report
print("Classification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
# Define stopwords
stop_words = set(stopwords.words('english'))
```

```python
# Clean text function
def clean_text(text):
    text = re.sub(r'\W+', ' ', text.lower())  # Remove non-alphanumeric characters and convert to lowercase
    return text
```

```python
# Filter toxic comments and clean text
toxic_comments = df_train[df_train['toxic'] == 1]['comment_text'].dropna().apply(clean_text)
```

```python
# Ensure all entries are strings
toxic_comments = toxic_comments.fillna("").astype(str)
```

```python
# Remove stopwords
toxic_comments_cleaned = toxic_comments.apply(
    lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
```

```python
# Combine all toxic comments into a single string
toxic_text = ' '.join(toxic_comments_cleaned)

# Generate Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(toxic_text)

# Display Word Cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud for Toxic Comments")
plt.show()
```
